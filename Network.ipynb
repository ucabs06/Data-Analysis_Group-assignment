```
import networkx as nx
import matplotlib.pyplot as plt
import operator
```

```
Data = open('Q5  Network/AZ_friends_network_above_100_friends.csv', "r")
Graphtype = nx.Graph()
G = nx.parse_edgelist(Data, delimiter=';', create_using=Graphtype,
                      nodetype=str)
```

```
def draw_power_law(G):

    print(list(G.edges()))
    print(G.degree(0))                             #返回某个节点的度
    print(G.degree())                                     #返回所有节点的度
    print(nx.degree_histogram(G))    #返回图中所有节点的度分布序列（从1至最大度的出现频次）

    degree = nx.degree_histogram(G)          #返回图中所有节点的度分布序列
    x = range(len(degree))[1::]                           #生成x轴序列，从1到最大度
    y = [z / float(sum(degree)) for z in degree][1::]
    print(len(degree))
    print(y)
    #将频次转换为频率，这用到Python的一个小技巧：列表内涵，Python的确很方便：）
    fig = plt.figure()
    plt.loglog(x,y,color="red",linewidth=0.5)           #在双对数坐标轴上绘制度分布曲线
    fig.suptitle('Power-law Distribution', fontsize=20)
    plt.xlabel('k', fontsize=18)
    plt.ylabel('P(k)', fontsize=16)
    plt.show()            #显示图表
```
